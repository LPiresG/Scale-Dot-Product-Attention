```
import numpy as np

class ScaleDotProductAttention:
    def __init__(self):
        pass

    def softmax(self, a):
        a = a - np.max(a, axis=1, keepdims=True)
        exp_a = np.exp(a)
        division = exp_a / np.sum(exp_a, axis=1, keepdims=True)
        return division

    def attentionFormula(self, Q, K, V):
        score = Q @ K.T
        dk = K.shape[1]
        scaling = np.sqrt(dk)
        scaled_score = score / scaling
        softmax_result = self.softmax(scaled_score)
        attention = softmax_result @ V
        return attention, softmax_result

```

```
Q = np.array([[1, 0], [0, 1]])
K = np.array([[1, 0], [0, 1]])
V = np.array([[1, 2], [3, 4]])
```

```
result = ScaleDotProductAttention()
attention, softmax_result = result.attentionFormula(Q, K, V)
print('Saida: \n', attention)
print('Resultado do softmax: \n', softmax_result)
```
